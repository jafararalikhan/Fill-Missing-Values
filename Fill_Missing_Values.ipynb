#Import Dependency

```python
import pandas as pd
```

#Load CSV File

```python
df = pd.read_csv('/content/missing.csv')
```

#Print DataFrame

```python
df
```

#Null Value Fill With Mod() From Deptment Column

```python
df['dept'] = df['dept'].fillna(df['dept'].mode()[0])
df['dept']
```

#Find outlier In salary Column

```python
df['salary'].plot(kind='box')
```

#Find Outlier In Age Column

```python
df['Age'].plot(kind='box')
```

#Null Value Fill With Mean() From Salary Column

```python
df['salary'] = df['salary'].fillna(df['salary'].mean())
df
```

#Null Value Fill With Median() From Age Column

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
df
```
